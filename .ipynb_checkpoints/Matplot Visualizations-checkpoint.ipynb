```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Plotting a Series

```python
ser1 = pd.Series(np.random.randn(50)).cumsum()
print(list(ser1))
ser1.plot(kind='hist')
```

```python
ser1.plot?
```

```python
# Figure size in inches

ser1.plot(kind='hist', figsize=(6,4))
```

```python
# If plot type is not mentioned then python will take the call by itself
ser1.plot(figsize=(8,4))
```

```python
# Chart with options
ser1.plot(kind = 'line',
          grid = True,
          #bins = 50,
          legend = True,
          label = 'TimeSeries',
          title = 'Random Normal Numbers - Cumulative Series',
          xlim = (0, 10), # used to eliminate the outlier values along x-axis
          ylim = (0, 4), # used to eliminate the outlier values along y-axis
          # xticks = np.arange(0, 100, 10),
          # yticks = np.arange(-10, 12, 2),
          # style = 'go-',
          # alpha = 0.2,
          figsize = (10,4),
          color = 'r'
         )
```

```python
# One variable as a histogram

pd.Series(np.random.randn(10000)).plot(kind='hist',bins=50, color='r', alpha=0.6, title='A histogram')
```

```python
# Another Histogram

pd.Series(np.random.randn(1000)).plot.hist(bins=50, color='r', alpha=0.4, title = "Another Histogram")
```

```python
# Plotting random int series count
rand_int_ser = pd.Series(pd.cut(np.random.randint(1,100, 1000), 5, labels=list('ABCDE'))).value_counts()

rand_int_ser.plot.bar(title = 'A Bar Chart', ylim = (0,400), alpha = 0.6)
                                                                                                   
```

```python
# Plottin beta values
s1 = pd.Series(np.random.beta(2,5,1000)),
s2 = pd.Series(np.random.beta(0.2, 0.5, 1000))
s1.plot.box()
```

```python
s2.hist(bins=30)
```

# Plotting a DataFrame

```python
df = pd.DataFrame(np.random.randn(5000,5), columns=list('ABCDE'))
df
```

```python
# Default Plot
df.cumsum().plot(figsize=(8,4), legend=True)
```

```python
# Each plot on its own plot

df.cumsum().plot(figsize=(10,6), subplots=True)
```
