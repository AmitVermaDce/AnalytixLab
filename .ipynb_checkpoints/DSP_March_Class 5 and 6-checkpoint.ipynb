## Outliers

**Values which are extremely high or extremely low**


```python
# mean +/- 3SD
import numpy as np
nd = np.random.normal(loc=190,scale=45,size=1000)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
nd = pd.Series(nd)
nd.hist(bins = 100)
```

```python
UC = nd.mean() + 3 * nd.std()
LC = nd.mean() - 3 * nd.std()
```

```python
nd.loc[(nd < LC) | (nd > UC)]
```

```python
# Capping - substituting
# > UC = UC
# < LC = LC
```

```python
# capping < LC with LC
nd1 = nd.clip_lower(LC) 
nd1.loc[nd1 == LC]
```

```python
# capping > UC with UC
nd1 = nd.clip_upper(UC) 
nd1.loc[nd1 == UC]
```

```python
# IQR
#UC = Q3 + 1.5 * IQR
#LC = Q1 - 1.5 * IQR
```

```python
# Quantiles :
# 0% 25% 50% 75% 100%
# min Q1 median Q3 max
# list takes numbers bn 0 and 1
# 0 means 0%
# 1 mean 100%
# 0, 20, 25, 50, 75, 90, 99, 100
# 0,0.2,0.25,0.5,0.75,0.9,0.95,0.99,1
nd.quantile([0,0.2,0.25,0.5,0.75,0.9,0.95,0.99,1])
```

```python
# IQR - Inter Quantile Range
# Q1 - 25%
# Q3 - 75%

# IQR = Q3 - Q1
```

```python
# DataFrame.groupby() + agg() or apply()

# sum, mean, std, var, min, max, length -> use agg()

# GrpTotal/GrandTotal * 100 -> apply()
```

```python
temp = stores.groupby("Location") # temp contains the grouping information, all the data and the column names

Res = temp.TotalSales.agg({"SumOfTotalSales":"sum"})

Res = Res.reset_index()

Res
```

```python
temp = stores.groupby(["Location"]) # temp contains the grouping information, all the data and the column names

Res = temp[["TotalSales"]].agg({"SumOfTotalSales":"sum"})

Res = Res.reset_index()

Res
```

```python
# Calculate Mean of OperatingCost wrt Location and StoreType

temp = stores.groupby(["Location","StoreType"])
Res2 = temp["OperatingCost"].agg({"AvgOfOperatingCost":"mean"})
Res2.reset_index(inplace = True)
Res2

stores.groupby(["Location","StoreType"]).OperatingCost.agg({"AvgOfOperatingCost":"mean"}).reset_index(inplace = True)
```

```python
def topN(data, col, N):
    """
    Takes a dataframe, and returns N rows for given column (sorted by this column desc)
    """
    #return data.sort_values(by=col, ascending=False).loc[:, col].head(N)
    return data.sort_values(by=col, ascending=False).head(N)
```
